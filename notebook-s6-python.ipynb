# Semana 6: Matplotlib y Seaborn

Bienvenidos a la clase sobre Matplotlib y Seaborn. Hoy vamos a aprender a crear y personalizar gráficos utilizando estas dos poderosas bibliotecas de Python.

```python
## Importación de Bibliotecas necesarias

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Matplotlib

## Crear un Gráfico Básico
Vamos a comenzar con un gráfico muy simple para entender la estructura básica.

**Gráfico de Líneas**

*Para qué se usa:*
- Visualizar tendencias a lo largo del tiempo o secuencias.
- Comparar cambios entre diferentes conjuntos de datos.

*Cuándo es mejor usarlo:*
- Cuando tienes datos continuos.
- Cuando quieres mostrar cambios en los datos a lo largo del tiempo.

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el gráfico
plt.plot(x, y)
plt.title('Gráfico de Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

- `plt.plot(x, y):` Dibuja un gráfico de líneas de x contra y.
- `plt.title`, `plt.xlabel`, `plt.ylabel`: Añaden el título y las etiquetas de los ejes.
- `plt.show()`: Muestra el gráfico.

#### Personalización del Gráfico de Líneas
Podemos personalizar el gráfico de líneas cambiando colores, estilos de línea, y añadiendo marcadores.

```python
plt.plot(x, y, color='red', linestyle=':', marker='d', markersize=5, label='Seno')
plt.title('Gráfico de Líneas Personalizado')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid(True)
plt.show()
```

**Parámetros**:

- `color`: Define el color de la línea (puede ser 'red', 'green', '#123456', etc.).
- `linestyle`: Define el estilo de la línea ('-', '--', '-.', ':').
- `marker`: Añade marcadores a los puntos de datos ('o', 's', '^', 'd', etc.).
- `markersize`: Define el tamaño de los marcadores.
- `label`: Añade una etiqueta para la leyenda.
- `plt.legend()`: Muestra la leyenda del gráfico.


## Otros Tipos de Gráficos

**Gráfico de Barra**

*Para qué se usa:*
- Comparar diferentes categorías.
- Mostrar valores discretos y comparar cantidades.

*Cuándo es mejor usarlo:*
- Cuando tienes datos categóricos.
- Cuando necesitas mostrar la comparación de datos individuales.

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

**Parámetros**:
- `color`: Define el color de las barras.
- `edgecolor`: Define el color del borde de las barras.


**Gráfico de Dispersión**

*Para qué se usa:*
- Mostrar la relación entre dos variables.
- Identificar patrones, tendencias y correlaciones.

*Cuándo es mejor usarlo:*
- Cuando quieres explorar la relación entre dos variables.
- Cuando tienes muchos puntos de datos y quieres identificar distribuciones.

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 1000
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.title('Gráfico de Dispersión')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.colorbar()  # Añade una barra de colores
plt.show()
```

**Parámetros:**

- `s`: Define el tamaño de los puntos.
- `c`: Define los colores de los puntos.
- `alpha`: Define la transparencia de los puntos (entre 0 y 1).
- `cmap`: Define el mapa de colores utilizado.

**Gráfico de Histograma**

*Para qué se usa:*
- Mostrar la distribución de un conjunto de datos.
- Ver la frecuencia de datos en intervalos específicos.

*Cuándo es mejor usarlo:*
- Cuando quieres ver cómo se distribuyen tus datos.
- Cuando necesitas identificar la forma, dispersión y sesgo de los datos.

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

**Parámetros:**

- `bins`: Define el número de contenedores (bins) del histograma.
- `color`: Define el color de las barras del histograma.
- `edgecolor`: Define el color del borde de las barras.

## Pintar Dos Funciones en un Mismo Gráfico

Puedes pintar dos (o más) funciones en un mismo gráfico para compararlas directamente.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='Seno')
plt.plot(x, y2, label='Coseno')
plt.title('Gráfico de Seno y Coseno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.grid(True)
plt.show()
```

## Gráficos Dinámicos

Para crear gráficos dinámicos que se actualizan automáticamente, podemos usar la función FuncAnimation de Matplotlib.

Importante:
- La visualización no es soportada por jupyter.
- Para guardarlo necesitas instalar ffmpeg.

```python
import matplotlib.animation as animation

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y)

# Función de actualización
def update(num, x, line):
    line.set_ydata(np.sin(x + num / 10.0))  # Actualiza los datos de y
    return line,

# Crear la animación y asignarla a una variable
ani = animation.FuncAnimation(fig, update, frames=100, fargs=[x, line], interval=50)

# Guardar la animación como un archivo de video
ani.save('seno_animacion.mp4', writer='ffmpeg', fps=30)

# Añadir los títulos y etiquetas
plt.title('Gráfico Dinámico de Seno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.grid(True)

# Mostrar el gráfico (opcional)
plt.show()
```

## Mejorar la Apariencia de los Gráficos

1. Cambiar el Tamaño de la Figura

Podemos cambiar el tamaño de la figura para que se ajuste mejor a nuestras necesidades.

```python
plt.figure(figsize=(12, 6)) #figsize: Define el tamaño de la figura en pulgadas (ancho, alto).
plt.plot(x, y)
plt.title('Gráfico de Líneas con Tamaño Ajustado')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

2. Añadir Subgráficos

Podemos añadir múltiples subgráficos en una sola figura.

```python
_ , axs = plt.subplots(2, 2, figsize=(10, 10))

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.log(x + 1)

axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Seno')

axs[0, 1].plot(x, y2, 'r')
axs[0, 1].set_title('Coseno')

axs[1, 0].plot(x, y3, 'g')
axs[1, 0].set_title('Tangente')

axs[1, 1].plot(x, y4, 'b')
axs[1, 1].set_title('Logaritmo')

for ax in axs.flat:
    ax.set(xlabel='Eje X', ylabel='Eje Y')

# Ajustar el espacio entre subgráficos
plt.tight_layout()
plt.show()
```

**Parámetros:**

- `plt.subplots`: Crea una cuadrícula de subgráficos.
- `figsize`: Define el tamaño de la figura.
- `axs.flat`: Permite iterar sobre todos los subgráficos.

## Guardar Figuras

Podemos guardar nuestras figuras en diferentes formatos de archivo.

```python
plt.plot(x, y)
plt.title('Gráfico de Líneas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.savefig('grafico_lineas.png')  # Guarda el gráfico como un archivo PNG
plt.savefig('grafico_lineas.pdf')  # Guarda el gráfico como un archivo PDF
plt.show()
```

## TODOS...o la mayoría de gráficos que puedes hacer


**Gráfico de Líneas (Line Plot)**

`plt.plot(x, y)`

- **Uso:** Visualizar tendencias a lo largo del tiempo o secuencias.
- **Aplicación:** Series temporales, datos continuos.

**Gráfico de Barras (Bar Plot)**

`plt.bar(categories, values)`

- **Uso:** Comparar diferentes categorías.
- **Aplicación:** Comparación de datos discretos como ventas por producto.

**Gráfico de Barras Apiladas (Stacked Bar Plot)**

`plt.bar(categories, values, bottom=values2)`

- **Uso:** Mostrar la contribución de cada parte a un total.
- **Aplicación:** Desglose de ventas por región y producto.

**Gráfico de Dispersión (Scatter Plot)**

`plt.scatter(x, y)`

- **Uso:** Mostrar la relación entre dos variables.
- **Aplicación:** Identificación de patrones, correlaciones.

**Histograma (Histogram)**

`plt.hist(data, bins=30)`

- **Uso:** Mostrar la distribución de un conjunto de datos.
- **Aplicación:** Análisis de frecuencia de datos continuos.

**Gráfico de Área (Area Plot)**

`plt.fill_between(x, y)`

- **Uso:** Similar al gráfico de líneas, pero el área bajo la línea está rellena.
- **Aplicación:** Visualización de volúmenes acumulados a lo largo del tiempo.

**Gráfico de Cajas (Box Plot)**

`plt.boxplot(data)`

- **Uso:** Mostrar la distribución de datos basada en un resumen de cinco números.
- **Aplicación:** Comparación de distribuciones entre diferentes categorías.

**Gráfico de Violín (Violin Plot)**

`plt.violinplot(data)`

- **Uso:** Combina el gráfico de caja y el gráfico de densidad.
- **Aplicación:** Visualización de la distribución de datos y su densidad.

**Gráfico de Pastel (Pie Chart)**

`plt.pie(values, labels=categories)`

- **Uso:** Mostrar proporciones de un total.
- **Aplicación:** Distribución porcentual de un conjunto de categorías.

**Gráfico de Tallo y Hoja (Stem Plot)**

`plt.stem(x, y)`

- **Uso:** Similar al gráfico de líneas pero con marcadores en los puntos de datos.
- **Aplicación:** Datos discretos que varían en una línea base.

**Gráfico de Contorno (Contour Plot)**

`plt.contour(X, Y, Z)`

- **Uso:** Mostrar líneas de igual valor en un espacio bidimensional.
- **Aplicación:** Visualización de superficies tridimensionales en dos dimensiones.

**Gráfico de Contorno Relleno (Filled Contour Plot)**

`plt.contourf(X, Y, Z)`

- **Uso:** Similar al gráfico de contorno, pero las áreas entre líneas están rellenas.
- **Aplicación:** Visualización de superficies con zonas coloreadas.

**Gráfico de Calor (Heatmap)**

`plt.imshow(data, cmap='hot')`

- **Uso:** Mostrar la intensidad o frecuencia de datos en una matriz bidimensional.
- **Aplicación:** Mapas de calor de correlaciones, distribución de datos.

**Gráfico de Hexbin (Hexbin Plot)**

`plt.hexbin(x, y, gridsize=30, cmap='Blues')`

- **Uso:** Similar al gráfico de dispersión, pero los puntos se agrupan en hexágonos.
- **Aplicación:** Densidad de puntos en grandes conjuntos de datos.

**Gráfico Polar (Polar Plot)**

`plt.polar(theta, r)`

- **Uso:** Visualizar datos en coordenadas polares.
- **Aplicación:** Datos cíclicos como direcciones de viento.

**Gráfico de Burbuja (Bubble Plot)**

`plt.scatter(x, y, s=sizes)`

- **Uso:** Similar al gráfico de dispersión, pero los puntos tienen tamaños variables.
- **Aplicación:** Visualización de tres dimensiones de datos en un gráfico 2D.

**Gráfico de Series Temporales (Time Series Plot)**

`plt.plot_date(dates, values)`

- **Uso:** Visualizar datos a lo largo del tiempo.
- **Aplicación:** Series temporales, análisis financiero.


# Seaborn

## ¿Por Qué Usar Seaborn?
Seaborn proporciona una interfaz de alto nivel para dibujar gráficos estadísticos, facilitando la creación de visualizaciones atractivas con menos código. Aquí te mostramos cómo Seaborn puede simplificar tu trabajo.

```python
sns.set(style="whitegrid")  # Estilo de fondo con cuadrícula blanca
plt.rcParams['figure.figsize'] = (10, 6)  # Tamaño de las figuras
```

## Gráficos Avanzados y Únicos con Seaborn

### Gráfico de Correlación (Correlation Plot)
El gráfico de correlación muestra la relación entre diferentes variables.

```python
# Datos de ejemplo
data = sns.load_dataset('iris')

# Crear el gráfico de correlación
corr = data.drop(columns=['species']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title('Mapa de Correlación')
plt.show()
```

**Parámetros**:

- `data`: Conjunto de datos utilizado.
- `annot`=True: Muestra los valores dentro de las celdas.
- `cmap`: Mapa de colores utilizado (por ejemplo, 'coolwarm').
- `center`=0: Centro del mapa de colores.

Muestra las relaciones entre todas las variables numéricas de un conjunto de datos.
Se usa en análisis exploratorio de datos, detección de correlaciones.

### Gráfico de Pares (Pair Plot)

El gráfico de pares muestra relaciones entre todas las variables de un conjunto de datos.

```python
# Crear el gráfico de pares
sns.pairplot(data, hue='species', markers=["o", "s", "D"])
plt.suptitle('Gráfico de Pares de Iris', y=1.02)
plt.show()
```

**Parámetros:**

- `data`: Conjunto de datos utilizado.
- `hue`: Variable categórica para colorear los gráficos.
- `markers`: Lista de marcadores para cada categoría.

Visualizar relaciones entre múltiples variables y su distribución. Se usa en análisis exploratorio de datos, identificación de patrones.

### Gráfico de Catplot
El catplot permite crear varios tipos de gráficos categóricos de manera flexible.

```python
# Crear un catplot
tips = sns.load_dataset('tips')
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips)
plt.title('Catplot de Total Bill por Día y Sexo')
plt.show()
```

**Parámetros**:
- `x`, `y`: Variables en los ejes x e y.
- `hue`: Variable categórica para colorear los gráficos.
- `kind`: Tipo de gráfico categórico ('swarm', 'box', 'violin', etc.).
- `data`: Conjunto de datos utilizado.

Se usa para visualizar la distribución de datos categóricos.
Aplicación en el análisis de datos categóricos, comparación entre grupos.

## Gráfico de Facetas (Facet Grid)
FacetGrid permite crear una cuadrícula de gráficos basados en la combinación de valores de una o más variables

```python
# Crear un FacetGrid
g = sns.FacetGrid(data=tips, col='time', row='sex', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.suptitle('FacetGrid de Total Bill vs Tip')
plt.show()
```

**Parámetros:**
- `data`: Conjunto de datos utilizado.
- `col`, `row`: Variables categóricas para crear las facetas.
- `margin_titles`: Mostrar títulos en los márgenes de las facetas.

Visualizar relaciones entre variables en subgrupos de datos.
Aplicación: Comparación entre subgrupos, análisis detallado de datos categóricos.

### Gráfico de Densidad Bivariada (Joint Plot)
El jointplot muestra la relación entre dos variables junto con las distribuciones marginales de cada variable.

```python
# Crear un jointplot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.suptitle('Joint Plot de Total Bill y Tip', y=1.02)
plt.show()
```

**Parámetros**:
- `x`, `y`: Variables en los ejes x e y.
- `data`: Conjunto de datos utilizado.
- `kind`: Tipo de gráfico ('scatter', 'kde', 'hex', etc.).
- `fill`: Rellenar las áreas bajo las curvas de densidad.

Visualizar relaciones bivariadas. Análisis exploratorio de datos, identificación de patrones y distribuciones.

### Gráfico de Strip Plot
El stripplot muestra la distribución de datos junto con puntos individuales.

```python
# Crear un stripplot
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
plt.title('Strip Plot de Total Bill por Día')
plt.show()
```

**Parámetros**:
- `x`, `y`: Variables en los ejes x e y.
- `data`: Conjunto de datos utilizado.
- `jitter`: Desplazar los puntos para evitar solapamientos.

Visualizar la distribución de datos categóricos junto con puntos individuales. Aplicación: Análisis de dispersión, identificación de valores atípicos.

## Gráfico de Relación (Rel Plot)
El relplot facilita la creación de gráficos de dispersión y gráficos de líneas con múltiples niveles de agrupació

```python
# Crear un relplot
sns.relplot(x='total_bill', y='tip', hue='smoker', col='time', data=tips)
plt.suptitle('Rel Plot de Total Bill y Tip por Smoker y Time', y=1.02)
plt.show()
```

**Parámetros**:
- `x`, `y`: Variables en los ejes x e y.
- `hue`: Variable categórica para colorear los gráficos.
- `col`: Variable categórica para crear columnas.
- `data`: Conjunto de datos utilizado.

Visualizar relaciones entre múltiples variables y grupos. Análisis exploratorio de datos, identificación de patrones y tendencias.
